### Desafio 01
Criar uma escala de valores em milhões

```python
dados_vendas['Valor_anuncio_milhoes'] = dados_vendas['Valor_anuncio_float'] / 1000000
```

```python
dados_vendas.describe()
```

### Desafio 02
Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

```python
sns.set()
sns.set_style('darkgrid', {'axes.facecolor': '0.9'} )
plt.figure(figsize=(12, 6))
ax = sns.histplot(
    data = dados_vendas, 
    x = 'Valor_anuncio_milhoes',
    kde = True, 
    bins = 'auto', 
    lw = 1, 
    edgecolor = '#0de74c',
    color = '#077527'
)

ax.set_title('Imóveis na cidade de São Paulo', size=18)
ax.set(xlabel='Valor do imóvel em milhões de reais', ylabel='Quantidade de imóveis à venda')

plt.xlim((-0.5, 10.5))
plt.show()
```

### Desafio 03
Preço do metro quadrado por bairro e plotar em um gráfico ideal

```python
dados_vendas['Valor_m2'] = dados_vendas['Valor_anuncio_float'] / dados_vendas['Metragem']
```

```python
dados_vendas.groupby('Bairro')['Valor_m2'].describe()
```

```python
df_top25_head = pd.Series()
df_top25_head = dados_vendas.groupby('Bairro')['Valor_m2'].mean().sort_values(ascending=False).head(25)
df_top25_head = df_top25_head.to_frame()
df_top25_head['Bairro'] = df_top25_head.index
df_top25_head
```

```python
sns.set()
sns.set_style('dark', {'axes.facecolor': '0.9'} )
plt.figure(figsize=(6, 10))
ax = sns.scatterplot(
    data = df_top25_head, 
    x = 'Valor_m2',
    y = 'Bairro',
    lw = 1, 
    edgecolor = '#0de74c',
    color = '#077527',
    s=50
    )

ax.set_title('Top 25 mais caros', size=18)
ax.set(xlabel='Valor do m² em R$', ylabel='Bairros')

plt.show()
```

Limitei a 25 para fazer uma visualização mais compacta. O mesmo método pode ser utilizado para plotar outras perspectiva, como os bairros mais baratos usando o `tail(25)`, por exemplo.
