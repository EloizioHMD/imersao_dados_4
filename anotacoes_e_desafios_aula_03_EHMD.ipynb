### Desafio 02
Tratar os outliers e comparar com os resultados. 


```python
dados_vendas.describe()
```

Primeiramente, parecisamos filtar o dados_vendas apenas para valores numéricos.

```python
dados_numerico = dados_vendas[['Metragem', 'Quartos', 'Banheiros', 'Vagas', 'Valor_anuncio_float', 'Valor_m2']]
```

Agora, vamos criar os elementos do boxplot em saparado, numa liguagem técnica as medidas separatrizes com objetivo de segregar os outliers.

```python
Q1_dados = dados_numerico.quantile(.25)
Q3_dados = dados_numerico.quantile(.75)
IIQ_dados = Q3_dados - Q1_dados
limite_inferior = Q1_dados - 1.5 * IIQ_dados
limite_superior = Q3_dados - 1.5 * IIQ_dados
```

Com a medidas separatrizes vamos filtrar o dataframe para eliminar os Outliers.

```python
filtro = (dados_numerico >= limite_inferior) & (dados_numerico <= limite_superior)
dados_filtro = dados_vendas[filtro]
```

Com os dados preparados poderemos plotar todos os gráficos separando com e sem outliers.

```python
sns.set(rc={'figure.figsize':(14,6)})

for i in dados_numerico:
  f, axes = plt.subplots(1, 2)
  f.suptitle(i + ' com Outliers e ' + i + ' sem Outliers')
  sns.boxplot(  x= i, data=dados_vendas, ax=axes[0])
  sns.boxplot(  x= i, data=dados_filtro, ax=axes[1])
  '\n'  

print()
```

Apesar do metódo não apresentar resultado de forma esperada, é perceptível que os ouliers impactam muito nos resultados. Assim, para a construção de uma modelo de predição adequado será necessário a limpesa desses dados.
